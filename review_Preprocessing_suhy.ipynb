### 데이터 전처리

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
review = pd.read_csv("olive_app_review_crawling_final.csv")
```

```python
review.columns = ["index", "date", "rate", "review", "like"]
```

```python
review = review.drop(['index'], axis=1)
```

```python
review['like'] = [re.sub(r'[^0-9]', '', x) for x in review.like]
```

```python
review = review.astype({"like":"int"})
```

```python
review['review'] = [re.sub(r'[]', '', x) for x in review.review]
review
```

```python
review['date'] = pd.to_datetime(review['date'])
```

```python
review
```

```python
review.to_csv('olive_app_review_preprocessing_ver2.csv', encoding="utf-8-sig")
```
