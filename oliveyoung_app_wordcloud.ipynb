```python
with open('words_2122.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS

stop_words = list(STOPWORDS)
stop_words = ["", "", "", "", "", "", "", ""]

word_cloud = WordCloud(stopwords = stop_words, font_path='malgun', width=800, height=500, scale=2.0, max_font_size=250, background_color='white')

word_cloud.generate(text)

import matplotlib.pyplot as plt
plt.figure(figsize=(11,9))
plt.imshow(word_cloud)
plt.axis('off')

# wordcloud  
plt.savefig('oliveyoung.png', bbox_inches='tight')
plt.show()
```
