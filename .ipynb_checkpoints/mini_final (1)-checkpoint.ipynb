```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import csv
import pandas as pd
import folium

driver=webdriver.Chrome("chromedriver")
driver.implicitly_wait(3)
driver.get("http://map.naver.com")
```

```python
#location list만들기: 거점 설정
location_list=['종각역','혜화역','독립문역']
loc1=location_list[0]
loc2=location_list[1]
loc3=location_list[2]
```

```python
#검색창에 loc 찍어서 위치 설정한 후, '카페' 탐색
element = driver.find_element(By.CSS_SELECTOR,'div.input_box > .input_search')
element.send_keys(loc3)
time.sleep(1)
element.send_keys(Keys.ENTER)
time.sleep(5)

element = driver.find_element(By.CSS_SELECTOR,'div.input_box > .input_search')
cafe = "카페"
element.clear()
element.send_keys(cafe)
time.sleep(3)
element.send_keys(Keys.ENTER)
```

```python
def switch(iframe):
    driver.switch_to.default_content()
    time.sleep(1)
    driver.switch_to.frame(iframe)
    time.sleep(1)

switch("searchIframe")

food_category=driver.find_elements(By.CSS_SELECTOR,"._1pC6a > span > a")[2]
food_category.send_keys(Keys.ENTER)
category_coffee=driver.find_elements(By.CSS_SELECTOR,"._1c7mP a")[0]
category_coffee.send_keys(Keys.ENTER)
time.sleep(1)
```

```python
#Iframe의 body 가져오기
body = driver.find_element(By.CSS_SELECTOR,"body")
body.click()

#스크롤 내리기
while True:
    cafe_name_prev = driver.find_elements(By.CSS_SELECTOR,"._3nCBm span.OXiLu")
    print("누르기전",len(cafe_name_prev))
    
    for i in range(5):
        body.send_keys(Keys.PAGE_DOWN)
    
        body.send_keys(Keys.PAGE_DOWN)

    cafe_name_curr = driver.find_elements(By.CSS_SELECTOR,"._3nCBm span.OXiLu")
    print("스크롤후",len(cafe_name_curr))
    
    if cafe_name_prev==cafe_name_curr:
        break
        
    cafe_name_prev=cafe_name_curr
```

```python
#카페 이름,주소,별점 끌어오기
cafeList = []
while True:
    num = 0;
    try:
        for i in range(len(cafe_name_curr)):

            temp =[]
            switch("searchIframe")
            cafe_name=driver.find_elements(By.CSS_SELECTOR,"._3nCBm span.OXiLu")[i]
            print(cafe_name.text)
            temp.append(cafe_name.text)
            cafe_name.click()

            switch("entryIframe")
            cafe_address = driver.find_elements(By.CSS_SELECTOR,"._1Gmk4 > ._2yqUQ")[0]
            temp.append(cafe_address.text)
            print(cafe_address.text)

            try:
                cafe_rate = driver.find_element(By.CSS_SELECTOR,"span._1Y6hi._1A8_M > em")
                temp.append(cafe_rate.text)
                print(cafe_rate.text)
            except Exception as e:
                temp.append("별점없음")
                print("별점없음")

            switch("searchIframe")
            matched = cafe_name.text.find("점")
            if matched != -1 :
                temp.append("프랜차이즈")
                print("프랜차이즈")
            else:
                temp.append("개인카페")
                print("개인카페")
            cafeList.append(temp)

            del cafe_name
            del cafe_address
        
    except IndexError as e:
        print("-")
        
    f = open('cafe_list.csv','w',encoding='utf-8-sig',newline='')
    
    csvWriter = csv.writer(f)
    if num==0 : 
        csvWriter.writerow(['카페명','주소','별점','카페구분'])
    for li in cafeList:
            csvWriter.writerow(li)
    num += 1
    f.close()
    
    switch("searchIframe")
    
    anchers = driver.find_elements(By.CSS_SELECTOR,"._2ky45 a")
    currPage = -1

    for i, ancher in enumerate(anchers):
        if "_5vmWW" in ancher.get_attribute("class"):
            currPage = i
            break

    if "_34lTS" in anchers[-1].get_attribute("class"):
        break
    
    next_page = driver.find_elements(By.CSS_SELECTOR,"._2ky45 > ._2bgjK")[1]
    next_page.click()
```

```python
# 종각역,혜화역,독립문역 위도 경도
latitude = [ 37.569361, 37.5828275 ,37.57441]
longitude = [ 126.9808763, 127.0016805 ,126.95775]

g_map = folium.Map(location=[latitude[0],longitude[0]],zoom_start=18)

g_map
```

```python
cafe_area_A = pd.read_csv('cafe_list_종각1 (1).csv')
cafe_area_A.head()
```

```python
cafe_area_A.reset_index(inplace=True)
for i in range(len(cafe_area_A)):
    marker01 = folium.Marker([cafe_area_A.loc[i]['위도'], cafe_area_A.loc[i]['경도']],
                       icon = folium.Icon(color='blue', icon='star'))
    marker01.add_to(g_map)
```

```python
franchise = cafe_area_A.loc[cafe_area_A['카페구분'].isin(['프랜차이즈'])]
print(len(franchise))
franchise.head()
```

```python
ind=cafe_area_A.loc[cafe_area_A['카페구분'].isin(['개인카페'])]
print(len(ind))
ind.head()
```

```python
#종각역 시각화
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[latitude[0], longitude[0]],
    zoom_start=15
)

coords_franchise = franchise[['위도', '경도']]
coords_ind = ind[['위도', '경도']]

marker_cluster = MarkerCluster().add_to(m)


for lat, long in zip(coords_ind['위도'], coords_ind['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="red")).add_to(marker_cluster)
for lat, long in zip(coords_franchise['위도'], coords_franchise['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```
